```python
from scipy.stats import kstest
```

```python
columns = morphometrics_df.columns.tolist()
features = columns[1:]
```

```python
ks_scores = np.zeros(len(features))

for i, feat in enumerate(features):
    
    test_features = morphometrics_df[feat].values
    
    stat, pval = kstest(test_features,cdf='norm')
    ks_scores[i] = stat
```

```python
len(features)
```

```python
np.flip(np.argsort(ks_scores))[:]
```

```python
np.flip(np.sort(ks_scores))
```

```python
# top features
top_features = []
for feat_i in np.flip(np.argsort(ks_scores))[:16]:
    top_features.append(features[feat_i])
```

```python
top_features
```

```python
top_morphos_df = morphometrics_df[['cell_name']+top_features]
```

```python
top_morphos_df
```

```python
# don't worry about features with infinite values
plot_features = [#'num_stems',
                 'soma_length',
                 'axon_init_angle',                 
                 'basal_dist_l',
                 'basal_dist_u',
                 'basal_length_u',
                 'basal_length_l',
                 'basal_mean_angle_u',
                 'basal_mean_angle_l',]
```

```python
morphometrics_df.columns.tolist()
```

```python
plot_morpho_df = morphometrics_df[['cell_name']+plot_features]
```

```python
plot_morpho_df['basal_mean_angle_l'] = np.abs(plot_morpho_df['basal_mean_angle_l'].values)
```

```python
fig, axs = plt.subplots(2,4,figsize=(19,10),sharex=True)
axs = axs.ravel()

for i, feature in enumerate(plot_features):
    
    bx = sns.barplot(x='cell_name',y=feature,
                 color='k',
                 data=plot_morpho_df,ax=axs[i])

    labels = plot_morpho_df.cell_name.tolist()
    axs[i].set_xticklabels(labels,rotation=90)
   
    
    if i!=7:
        axs[i].set_xlabel('')
    else:
        axs[i].set_xlabel('Neuron-Type',size='xx-large')
        
    axs[i].set_ylabel(feature,size='x-large')
    
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.2)

plt.suptitle('Gross Morphological Properties',size='xx-large')
plt.tight_layout();
```
